# Writing Mermaid diagrams as HTML
This notebook describes how to write HTML pages that show
Mermaid diagrams.

## Modify your code
Below are changes needed to output the diagram

### 1. Add required imports
Add the imports below to the Python module from which you want
to generate the HTML page.

````python
from idaes_connectivity import Mermaid, Connectivity
````

### 2. Add function to create the page
It is probably most convenient to add a function like the following
to add the diagram in one function call.

This function creates the Connectivity from the model, then creates
the Mermaid diagram, then writes the diagram to a file of your choice.

````python
def create_diagram_html(model):
    # create connectivity from the model
    conn = Connectivity(input_model=model)
    # create the mermaid diagram from the connectivity
    mermaid_diagram = Mermaid(conn, component_images=True)
    # write out an HTML page with the diagram embedded;
    # you can use any filename here, "diagram.html" is not special
    mermaid_diagram.write_html("diagram.html")
````

### (3. Build Model)

To make this example self-contained, we build the model here.
Your code already does this, of course.

````python
# the code in this cell requires a newer version of IDAES,
# but any way you get to a built model will suffice
from idaes.core.util.structfs.tests import hda_flowsheet
model = hda_flowsheet.build()
````

### 4. Create the diagram
Call the helper function you created above to create the model.
In this case, it will be created in the file 'diagram.html'

````python
create_diagram_html(model)
````

## View the diagram
### 1. Start the image server (optional)
The Mermaid diagram can use SVG images for some common unit types.
However, this requires that a little local server is running.
There is a utility command that helps with this, `idaes-conn`.

There are actually two things you need to do. First, you need to copy
the images (once) to the standard location under `~/.idaes/img` by running this command:

```
idaes-conn --copy-images
```

You only have to run that command once after installing the package.

You will need to re-run the image server after you login. To get the custom images in the diagram, start a new terminal and run:
```
idaes-conn --start-image-server
```
**You only need to do this once.** You can do it at any time before you load the generated HTML page in the browser. This server must stay running, so don't kill the terminal session until you're done viewing the diagram.

Also, of course, if you do not care about the SVG images and are happy to just have boxes for each unit, you don't need to do anything in this step at all.

### 2. View the page in the browser
Open the file 'diagram.html' in a web browser.
Note that once you loaded it once into the browser once, you can write
a new HTML diagram and view the updated one by refreshing the page.

## Rinse and Repeat
Now you can rerun your build step as many times as you want,
each time calling your `create_diagram_html()` function to create
the diagram, and simply use the browser's reload button to view 
the updated results.
